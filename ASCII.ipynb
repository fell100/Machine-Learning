```python
def get_character(intensity):
  ascii_string = "$@B%8&WM#*oahkbdpqwmZO0QLCJUYXzcvunxrjft/\|()1{}[]?-_+~<>i!lI;:,^`'. "
  ascii_string = list(ascii_string)[::-1]
  return ascii_string[int(int( np.round(int(intensity) * int(len(ascii_string))/255) ) -1)]
```

```python
keth = cv2.GaussianBlur(keth,(35,35),0)
keth = cv2.resize(keth, (100,100))[:78]
plt.imshow(keth, cmap='gray')
```

```python
keth.shape
```

```python
keth = keth.astype('U')
```

```python
def ascii_img(img):
  W, H = img.shape 
  for v in range(0,H): 
      for u in range(0,W):
          img[u,v] = get_character(img[u,v])
  return img
```

```python
keth[0][:]
```

```python
ascii_keth = ascii_img(keth)
```

```python
with open("test.txt",'w',encoding = 'utf-8') as f:
  h, w = keth.shape
  x = ""
  for j in range(0,h):
    for i in ascii_keth[j][:]:
      x+= i
    f.write(f"{x}\n")
    x = ""
```

```python
mae1 = cv2.imread("/content/drive/MyDrive/HistogramasPDI/imagens/mae.jpg", cv2.IMREAD_GRAYSCALE)
mae2 = cv2.imread("/content/drive/MyDrive/HistogramasPDI/imagens/mae2.jpg", cv2.IMREAD_GRAYSCALE)
irma = cv2.imread("/content/drive/MyDrive/HistogramasPDI/imagens/irma.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
mae1 = cv2.GaussianBlur(mae1,(35,35),0)
mae1 = cv2.resize(mae1, (200,200))[:100]
plt.imshow(mae1, cmap='gray')
```

```python
mae2 = cv2.GaussianBlur(mae2,(15,15),0)
mae2 = cv2.resize(mae2, (350,350))[:200]
plt.imshow(mae2, cmap='gray')
```

```python
irma = cv2.GaussianBlur(irma,(25,25),0)
irma = cv2.resize(irma, (500,500))[50:250]
plt.imshow(irma, cmap='gray')
```

```python
def gera_txt(img_ascii, name):
  img_ascii = img_ascii.astype('U')
  img_ascii = ascii_img(img_ascii)

  with open(f"{name}.txt",'w',encoding = 'utf-8') as f:
    h, w = img_ascii.shape
    x = ""
    for j in range(0,h):
      for i in img_ascii[j][:]:
        x+= i
      f.write(f"{x}\n")
      x = ""
```

```python
gera_txt(mae1, "mae1")
gera_txt(mae2, "mae2")
gera_txt(irma, "irma")
```
